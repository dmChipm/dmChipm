## Problem 3 Answer (edit this cell)
1. (Describe the pattern)

The pattern shows an increase in usefulness starting in the summer months and peaks sometime around the months of January, then decreases in usefulness until the summer months of May, June, and July.

1. (Speculate why this pattern might exist)

As the seasons begin to turn cold Canadians tend have more time on their hands because there is less appeal to be outside and therefore more care and attention is put into their reviews.
```
# create 'ym' column

reviews['ym'] = reviews_beforeandafter.date.dt.to_period('M')

```

```
# check that you've created it correctly by finding the unique values of ym
# select the ym column and use the .unique() method

reviews.ym.unique()
```

```
# create the plot using a groupby summary + .plot method
# refer to notes for examples


temp = reviews.groupby('ym')['useful'].sum().reset_index()
temp

reviews.groupby(['ym'])['useful'].mean().plot()
```
